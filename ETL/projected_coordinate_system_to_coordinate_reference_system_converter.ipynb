# Projected  System to Coordinate Reference System Converter

Takes in data shapefiles with geometry fields using projected coordinate system (PCS) reference system and converts that data to coordinate reference system (CRS), then saves that as a separate shapefile to be used for future geospatial analysis. 

```python
import os
import pandas as pd
import geopandas as gpd
```

```python
county_fips = 453
election_year = '22'
election_type = 'g'
```

```python
shapefiles_filepath = os.path.join('..', 'data', 'geospatial_files', 'shapefiles', 'texas_legislative_council', f'20{election_year}/', f'vtds_{election_year}{election_type}'.upper())
# creates GeoDataFrame from Shapefiles
gdf = gpd.read_file(shapefiles_filepath.upper() + '.shp').set_crs(epsg=4326)
```

```python
gdf
```

```python
gdf['geometry_obj'] = gdf.geometry.astype(str)
```

```python
test_geometry_list = []
test_geometry = gdf.geometry_obj.iloc[0]
test_geometry = test_geometry.replace('POLYGON ', '').replace('(', '').replace(')', '')
test_geometry = test_geometry.split(',')
for geometry in test_geometry:
    geometry = geometry.split(' ')
    test_geometry_list.append(geometry[-2:])
test_geometry_list
```
